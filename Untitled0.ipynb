```python
# variáveis do tipo string
nome = 'João da Silva'
cidade = 'São Paulo'
cpf = '123.456.789-00'
print('nome'.upper())
print('cidade'.upper())
print('cpf'.upper())
print('nome'.lower())
print('cidade'.lower())
print('cpf'.lower())

posicao_nome = nome.find('ã')
posicao_cidade = cidade.find('ã')
posicao_cpf = cpf.find('ã')

print(f'Posição do caractere "ã" em nome: {posicao_nome}')
print(f'Posição do caractere "ã" em cidade: {posicao_cidade}')
print(f'Posição do caractere "ã" em CPF: {posicao_cpf}')

```

```python
# Variáveis do tipo string
nome = 'João da Silva'
cidade = 'São Paulo'
cpf = '123.456.789-00'

variaveis = [nome, cidade, cpf]
for var in variaveis:
    posicao = 0
    while 'ã' in var[posicao:]:
        posicao = var.find('ã', posicao)
        print(f'Posição do caractere "ã" em {var}: {posicao}')
        posicao += 1

```

```python
cpf = '123.456.789-00'
cpf_numerico = ''.join(filter(str.isdigit, cpf))
resultado = int(cpf_numerico)

print(f'Resultado da conversão: {resultado}')
```
